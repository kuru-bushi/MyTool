```python
import matplotlib.pyplot as plt
import numpy as np

x = np.arange(-5, 5, 0.01)
y = 3*x**3 - 4+x**2 + 2*x*1
plt.plot(x, y)
plt.show()
```

# シグモイド関数のプロット

```python
def sigmoid(x):
    return 1/(1+np.exp(-x))

x = np.arange(-5, 5, 0.01)
y = sigmoid(x)
plt.plot(x, y)
plt.show()
```

# step ステップ関数のプロット

```python
def step(x):
    return np.array(x > 0, dtype = np.int)

x = np.arange(-5, 5, 0.01)
y = step(x)
plt.plot(x, y)
plt.show()
```

```python
def ReLU(x):
    return np.maximum(0, x)

x = np.arange(-5, 5, 0.01)
y = ReLU(x)
plt.plot(x, y)
plt.show()
```

```python
# 勾配降下法
def gradient_descent(start, eta, iter):
    x = start
    for i in range(iter):
        x = x - eta*(2*x)
    return x
gradient_descent(start=4, eta = 0.1, iter = 100)
```

```python
# ポアソン分布のシミュレーション
x = np.random.poisson(size = 10**5, lam = 5.0)
x, height = np.unique(x, return_counts = True)
plt.bar(x = x, height = height/height.sum())
plt.show()
```

```python
# シミュレーション
sim = []

for i in range(10**5):
    sim.append(0)
    for j in range(7):
        sim[i] += np.random.poisson(size = 1, lam = 5.0)

x, height = np.unique(sim, return_counts = True)
plt.bar(x = x, height = height/height.sum())
plt.show()

```
